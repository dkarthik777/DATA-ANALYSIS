# **Conclusion**

Based on the analysis of Uber ride request data:

High Demand Periods (especially 5 AMâ€“9 AM and 5 PMâ€“9 PM) experience a significant supply-demand gap, where a large number of ride requests are either cancelled or marked as No Cars Available.

Airport pickup location suffers more from unfulfilled requests compared to the City, indicating poor supply during high outbound/inbound traveler times.

A large percentage of requests have no drivers assigned, especially during peak hours, revealing a driver shortage problem in certain time windows.

Most Trip Completed statuses are skewed toward mid-day hours when demand and supply are relatively balanced.
