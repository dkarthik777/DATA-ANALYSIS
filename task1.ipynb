```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data1 = pd.read_csv(r'C:\Users\dines\Downloads\credits.csv\credits.csv')
data2 = pd.read_csv(r'C:\Users\dines\Downloads\titles.csv\titles.csv')
```

```python
data.dropna(subset=['description'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data['seasons'].fillna(0, inplace=True)
```

```python
round(data['imdb_score'].mean(), 1)
```

```python
data.shape
```

```python
data['imdb_votes'].fillna(0, inplace=True)
```

```python
data.isnull().sum()
```

```python
round(data['tmdb_popularity'].mean(), 2)
data['tmdb_popularity'].fillna(round(data['tmdb_popularity'].mean(), 2), inplace=True)
```

```python
round(data['tmdb_score'].mean(), 2)
data['tmdb_score'].fillna(round(data['tmdb_score'].mean(), 2), inplace=True)
```

```python
data['character'].fillna('Unknown', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='tmdb_score')
plt.title('Boxplot of TMDB Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='release_year', bins=30, color= 'blue', kde=True)
plt.title('Boxplot of TMDB Score')
plt.show()
```

```python
sns.barplot(data=data, x='release_year', y='tmdb_score', color='blue')
plt.title('Average TMDB Score by Release Year')
plt.ylabel('Average TMDB Score')
plt.show()
```

```python
sns.scatterplot(data=data, x='imdb_votes', y='tmdb_score', alpha=0.5)
plt.title('imdb_votes vs TMDB Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='release_year', y='runtime', marker='o')
plt.title('runtime vs Release Year')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='type', y='tmdb_score', inner='quartile')
plt.title('Violin Plot of TMDB Score by type')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```
